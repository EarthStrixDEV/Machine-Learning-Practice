```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Score': [85, 90, np.nan, 78, 92], # Charlie ไม่มีคะแนน
    'Grade': ['A', 'A', 'B', 'C', np.nan], # Eve ไม่มีเกรด
    'Attendance': [95, 98, 80, np.nan, 90] # David ขาดการเข้าเรียนบางส่วน
}
data
```

```python
df_students = pd.DataFrame(data)
df_students
```

```python
df_students.isnull()
```

```python
df_students.isnull().sum()
```

```python
df_students.isnull().sum().sum()
```

```python
df_students.notnull()
```

```python
df_students.dropna()
```

```python
df_students.dropna(axis=1)
```

```python
df_all_nan_row = pd.DataFrame({
    'A': [1, 2, np.nan],
    'B': [3, 4, np.nan],
    'C': [np.nan, np.nan, np.nan]
})
df_all_nan_row
```

```python
df_all_nan_row.dropna(how='all')
```

```python
mean_score = df_students['Score'].mean()
df_students['Score'] = df_students['Score'].fillna(mean_score)
df_students
```

```python
df_students['Grade'].fillna('Unknown' ,inplace=True)
df_students
```

```python
df_students['Attendance'].fillna(method='ffill' ,inplace=True)
df_students
```
